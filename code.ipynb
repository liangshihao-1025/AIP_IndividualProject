```python

x=input("Enter first number  ")
y=input("Enter second number  ")




def number_type(number):
    try:
        int(number)
        return int(number) 
        
    except ValueError:
        pass

    try:
        float(number)
        return float(number)   
        
    except ValueError:
          print('Input error')
    again()
              

num_1 = number_type(x)
num_2 = number_type(y)
print(type(num_1), type(num_2))




def divide(num_1,num_2):
    if num_2 == 0:
        print('Error')
    else:
        return num_1 / num_2


    
def calculate(num_1, num_2, operation):
    if operation == '+':    #addition
        total = num_1 + num_2
        result = "Result is {} .".format(total)
    elif operation=='-':    #subtraction 
        total = num_1 - num_2
        result = "Result is {} .".format(total)
    elif operation=='*':    #multiplication
        total = num_1 * num_2
        result = "Result is {} .".format(total)
    elif operation=='/':    #division
        total = divide(num_1,num_2)
        result = "Result is {} .".format(total)
    else:
        result = "wrong"
        

    print(result)
    again()   


def again():
    calc_again = input('''Do you want to run again?
Please input Y for yes N or nothing for no.
    ''')
    
    if calc_again.upper() == 'Y':
        x=input("Enter first number  ")
        y=input("Enter second number  ")
        num_1 = number_type(x)
        num_2 = number_type(y)
        print(type(num_1), type(num_2))
        operation = input('''Please type the operation that you would like to complete (+, -, * or /): 
''')
        calculate(num_1, num_2, operation)
    else:
        print('See you later.')


    
    
operation = input('''Please type the operation that you would like to complete (+, -, * or /): 
''')
calculate(num_1, num_2, operation)
```
